```python
import numpy as np
import pandas as pd
```

### 1) Tum ogrencilerin bilgileri:

```python
py_mind = pd.ExcelFile("py_mind.xlsx")
py_sense = pd.ExcelFile("py_sense.xlsx")
py_science = pd.ExcelFile("py_science.xlsx")
py_opinion = pd.ExcelFile("py_opinion.xlsx")
```

```python
dosyalar = [py_mind, py_sense, py_science, py_opinion]
siniflar=[]
isim=[]
dogru=[]
yanlis=[]
bos=[]

for i in dosyalar:                                     # dort farklı excel dosyası için döngü.
    for j in range(len(i.sheet_names)):                # her dosyanın içindeki sayfa sayısı kadar döngü.
        
        data_mind = pd.read_excel(i, i.sheet_names[j]) # her sayfaya tek tek ulaşma.
        
        isim += [data_mind.columns[0]]                 # istenilen bilgiler hangi satir hangi sütunda ise onları alma.
        dogru += [data_mind.iloc[20,1]]
        yanlis += [data_mind.iloc[21,1]]
        bos += [data_mind.iloc[22,1]]
        
        if i == py_mind:
            sinif = "py_mind"
        elif i == py_sense:
            sinif = "py_sense"
        elif i == py_science:
            sinif = "py_science"
        elif i == py_opinion:
            sinif = "py_opinion"
        
        siniflar += [sinif]

all_students = pd.DataFrame({"Sinif ismi": siniflar, "Ogrenci Adi": isim ,"Dogru Sayisi": dogru, 
                             "Yanlis Sayisi": yanlis, "Bos Sayisi": bos})

all_students.index = all_students.index + 1            # indexleri sıfır yerine bir'den başlatma.
all_students 
```

### 2) Sinava giren ogrenci sayisi:

```python
number_of_students = all_students.index[-1]
number_of_students
```

### 3) Siniflarin ayri ayri dogru yanlis bos ortalamalari:

```python
class1 = all_students[all_students["Sinif ismi"] == "py_mind"]
class2 = all_students[all_students["Sinif ismi"] == "py_sense"]
class3 = all_students[all_students["Sinif ismi"] == "py_science"]
class4 = all_students[all_students["Sinif ismi"] == "py_opinion"]

# tüm ögrecileri sinif sinif ayirdik.
```

```python
print("py_mind average:\n", class1.mean(), "\n\npy_sense average:\n", class2.mean(), "\n\npy_science average:\n", class3.mean(),
      "\n\npy_opinion average:\n", class4.mean())
```

### 4) Her sinifin ayri ayri en basarili ogrencisi:

```python
winner_names1 = list(class1[class1["Dogru Sayisi"] == class1["Dogru Sayisi"].max()]["Ogrenci Adi"]) 
winner_names2 = list(class2[class2["Dogru Sayisi"] == class2["Dogru Sayisi"].max()]["Ogrenci Adi"]) 
winner_names3 = list(class3[class3["Dogru Sayisi"] == class3["Dogru Sayisi"].max()]["Ogrenci Adi"]) 
winner_names4 = list(class4[class4["Dogru Sayisi"] == class4["Dogru Sayisi"].max()]["Ogrenci Adi"]) 

# bu sekilde liste yapmak, "max" yapan birden fazla kişi varsa hepsinin ismini listeye almayı saglıyor.
```

```python
winners = winner_names1 + winner_names2 + winner_names3 + winner_names4
winners

# en yüksek doğru sayısı (net sayısı değil) olanların ismine bu şekilde ulaşılabiliyor. 
# asagida nete gore bir tablo yapmak istersek:
```

```python
cls_name=[]
point=[]

for i in winners:        
    
    data = all_students[all_students["Ogrenci Adi"] == i]
    cls_name.append(list(data["Sinif ismi"]))
    point.append(list(data["Dogru Sayisi"]-data["Yanlis Sayisi"]/4))
        
cls_name = [j for i in cls_name for j in i]                     # isimleri ve netleri liste tipinde listelere ekleyebildim.
point=[j for i in point for j in i]                             # o yüzden görüntüyü bu sekilde kurtarmaya calistim.

winners_df = pd.DataFrame({"SINIF":cls_name, "ISIM":winners, "NET":point}, index=list(range(1,len(winners)+1)))
winners_df
```

### 5) Tum sinavin en basarili ilk 3 ogrencisi:

```python
all_students["Net Sayisi"] = all_students["Dogru Sayisi"] - all_students["Yanlis Sayisi"]/4
# Net sayisi için yeni bir sütun oluşturup, sonra net sayisina göre siraliyoruz.
top_three = all_students.sort_values(by="Net Sayisi", ascending=False).head(3)
top_three.index = [1,2,3]
# Yeni index atayıp tüm satırların istediğimiz sütunlarını alıyoruz.
top_three.loc[:,["Sinif ismi", "Ogrenci Adi", "Net Sayisi"]]
```

### 6) Siniflarin standart sapmalari:

```python
std_deviation = [class1["Dogru Sayisi"].std(), class2["Dogru Sayisi"].std(), 
                 class3["Dogru Sayisi"].std(), class4["Dogru Sayisi"].std()]

class_name = [class1.iloc[0,0].upper(), class2.iloc[0,0].upper(), class3.iloc[0,0].upper(), class4.iloc[0,0].upper()]

std_df = pd.DataFrame({"STD.SAPMA": std_deviation}, index=class_name)
std_df
```

### 7) Sinavin dogru yanlis ve bos ortalamalari:

```python
avg_df = pd.DataFrame({"DOGRU": [all_students["Dogru Sayisi"].mean()], 
                       "YANLIS": [all_students["Yanlis Sayisi"].mean()],
                       "BOS": [all_students["Bos Sayisi"].mean()]},index=["ORTALAMA"])
avg_df
```
